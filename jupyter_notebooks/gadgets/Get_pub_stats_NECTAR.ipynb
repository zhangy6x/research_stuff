```python
# ris_keys = ['year','type_of_reference', 'title', 'journal_name',]
year_list = []
type_list = []
title_list = []
journal_list = []
df = pd.DataFrame()
```

```python
with open(file_path, 'r') as bibliography_file:
    entries = readris(bibliography_file)
for entry in entries:
    year_list.append(entry['year'])
    type_list.append(entry['type_of_reference'])
    title_list.append(entry['title'])
    journal_list.append(entry['journal_name'])
    
df['year'] = year_list
df['type_of_reference'] = type_list
df['title'] = title_list
df['journal_name'] = journal_list

```
