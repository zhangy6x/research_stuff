```python
pp = p.T
df1 = pd.DataFrame()
for i, col in enumerate(pp):
    df1[i] = col
col_name_1 = ['E_eV', 'I_angstrom', 'f(E)', 's(E)', 'f(I)', 's(I)']
df1.columns = col_name_1
df1
```

```python
qq = q.T
df2 = pd.DataFrame()
for i, col in enumerate(qq):
    df2[i] = col
col_name_2 = ['t_us', 'E_eV', 'f', 's']
df2.columns = col_name_2
df2
```

```python
# df1.to_clipboard()
df1.to_csv('section_1.csv', index=False)
```

```python
energy_list = set(list(df2['E_eV']))
```

```python
len(energy_list)
```

```python
qqq = {}
for index, each_energy in enumerate(energy_list):
    t_us = []
    e_ev = []
    f = []
    s = []
    df = pd.DataFrame()
    for each_line in q:
        if each_energy == each_line[1]:
            t_us.append(each_line[0])
            e_ev.append(each_line[1])
            f.append(each_line[2])
            s.append(each_line[3])
    df['t_us'] = t_us
    df['E_eV'] = e_ev
    df['f'] = f
    df['s'] = s
    file_name = 'energy_' + str(index+1) + '.csv'
    df.to_csv(file_name, index=False)
    qqq[each_energy] = df
```
