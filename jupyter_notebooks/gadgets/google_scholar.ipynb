```python
pub_title_list = []
for idx, pub in enumerate(author['publications']):
    if int(pub['bib']['pub_year'])>=after_year:
        print(pub['bib']['pub_year'], pub['bib']['title'])
        pub_title_list.append(pub['bib']['title'])
pub_title_list
```

```python
paper_search_query = scholarly.search_single_pub(pub_title_list[2], filled=True)
pprint(paper_search_query)
```

```python
author_df = pd.DataFrame()
author_name_list = []
author_name_raw_list = []
last_name_list = []
affiliation_list = []
year_list = []
first_name_list = []
title_list = []
affiliation_dict = {}
for each_pub in pub_title_list:
    print("Filling...", each_pub)
    _paper_search_query = scholarly.search_single_pub(each_pub, filled=True)
    _author_id_list = _paper_search_query['author_id']
    _author_list = _paper_search_query['bib']['author'].split(' and ')
    for _author_id in _author_id_list:
        if len(_author_id) >1:
            _author_search_query = scholarly.search_author_id(_author_id)
            _affiliation = _author_search_query['affiliation']
            _name = _author_search_query['name']
            affiliation_dict[_name] = _affiliation
    for each_author in _author_list:
        _first_last_name = each_author.split(', ')
        last_name_list.append(_first_last_name[0])
        first_name_list.append(_first_last_name[-1])
        _author_name = _first_last_name[-1] + ' ' + _first_last_name[0]
        author_name_list.append(_author_name)
        author_name_raw_list.append(each_author)
        if _author_name in affiliation_dict.keys():
            affiliation_list.append(affiliation_dict[_author_name])
        else:
            affiliation_list.append('')
        year_list.append(_paper_search_query['bib']['pub_year'])
        title_list.append(each_pub)
    
author_df['raw_name'] = author_name_raw_list
author_df['name'] = author_name_list
author_df['last_name'] = last_name_list
author_df['first_name'] = first_name_list
author_df['affiliation'] = affiliation_list
author_df['year'] = year_list
author_df['title'] = title_list
```
