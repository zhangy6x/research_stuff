```python
import os
import numpy as np
import sys
import pandas as pd
root_folder = os.path.dirname(os.getcwd())
sys.path.append(root_folder)
```

```python
###################### need to change start
ipts="26032"
###################### need to change end
scan_loc = "/HFIR/CG1D/IPTS-"+ipts+"/shared/analysis"
scan_list = os.listdir(scan_loc)
sorted(scan_list, reverse=True)
```

```python
scan_name = "2023_05_28_powder_fine"
#########################################
scan_dir = os.path.join(scan_loc, scan_name)
file_list = os.listdir(scan_dir)
file_list = sorted(file_list, reverse=False)
file_list
```

```python
src_list = []
dst_list = []
for each_file in file_list:
    for each_type in ['DFI_','DPC_', 'TI_']:
        if each_type in each_file:
            _str_list = each_file.split('_')
            _str_list[1] = 'ACL'
            _pos_str = int(_str_list[2].split('.')[0])
            _str_list[2] = "{:03d}".format(_pos_str) + '.fits'
            src_list.append(each_file)
            dst_list.append('_'.join(_str_list))
```

```python
for idx, each in enumerate(src_list):
    _src_file = os.path.join(scan_dir, each)
    _dst_file = os.path.join(scan_dir, dst_list[idx])
    os.rename(_src_file, _dst_file)
```
