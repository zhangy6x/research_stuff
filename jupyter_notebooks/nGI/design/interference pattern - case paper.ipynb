```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def psi(x, z, An, n_list, d, wavelength):
    L_T = d*d/wavelength
    rt = 0.
    k_d = 2*np.pi/d
    for n in n_list:
        A = An(n)
        rt += A*np.exp(1j*n*k_d*x) * np.exp(-1j*n*n*np.pi*(z/L_T))
        continue
    return rt
```

```python
def make_An(f):
    def An(n):
        if n==0: return f
        return np.sin(n*np.pi*f)/n/np.pi
    return An
```

```python
d=1e-5
wavelength=4e-10
xs=np.arange(-2e-5, 2e-5, .5e-7)
L_T = d*d/wavelength
zs = np.arange(0, 1.7*L_T, 0.002*L_T)
An = make_An(0.1)
n_list = range(-25, 26)
Is = np.array([np.abs(psi(xs,z,An,n_list,d,wavelength))**2 for z in zs])

plt.figure()
zg, xg = np.meshgrid(zs/L_T, xs)
plt.pcolormesh(zg, xg, Is.T)
plt.clim(0, 0.5)
```
